# Data Quality Checks

```python
df.isnull().sum()
```

```python
df.isnull().mean()*100
```

```python
df.duplicated().sum()
```

```python
df['hotel'].unique()

df['hotel'].value_counts()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['company'] = df['company'].fillna(0)
```

```python
df['agent'] = df['agent'].fillna(0)
```

```python
df['country'] = df['country'].fillna(df['country'].mode()[0])
```

```python
df['children'] = df['children'].fillna(df['children'].median())
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['lead_time'].describe()
```

```python
df['lead_time'].skew()
```

```python
df['hotel'].value_counts()

df['hotel'].value_counts(normalize=True)
```

```python
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
```

```python
plt.figure(figsize=(10,6))

sns.histplot(df['lead_time'], bins=40, kde=True)

plt.title("Distribution of Lead Time")
plt.xlabel("Lead Time")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df['lead_time'])

plt.title("Boxplot of Lead Time")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df, x='hotel')

plt.title("Number of Bookings by Hotel Type")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df, x='is_canceled')

plt.title("Booking Cancellation Distribution")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(data=df,
              x='hotel',
              hue='is_canceled')

plt.title("Cancellation by Hotel Type")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(data=df,
            x='hotel',
            y='adr')

plt.title("Average Daily Rate by Hotel")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(data=df,
                x='lead_time',
                y='adr')

plt.title("Lead Time vs ADR")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.violinplot(data=df,
               x='hotel',
               y='adr')

plt.title("ADR Distribution by Hotel")

plt.show()
```

```python
plt.figure(figsize=(14,10))

corr = df.select_dtypes(include='number').corr()

sns.heatmap(corr,
            cmap='coolwarm',
            annot=True,
            fmt=".2f")

plt.title("Correlation Matrix")

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.countplot(data=df,
              x='arrival_date_month',
              order=df['arrival_date_month'].value_counts().index)

plt.xticks(rotation=45)

plt.title("Bookings by Month")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df,
              x='customer_type')

plt.xticks(rotation=20)

plt.title("Customer Types")

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.countplot(data=df,
              x='market_segment')

plt.xticks(rotation=45)

plt.title("Market Segment Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df,
              x='deposit_type')

plt.title("Deposit Type")

plt.show()
```

```python
top_country = df['country'].value_counts().head(10)

plt.figure(figsize=(10,6))

sns.barplot(x=top_country.index,
            y=top_country.values)

plt.title("Top 10 Countries")

plt.xlabel("Country")
plt.ylabel("Bookings")

plt.show()
```

```python
cols = ['lead_time',
        'adr',
        'stays_in_week_nights',
        'stays_in_weekend_nights',
        'is_canceled']

sns.pairplot(df[cols])

plt.show()
```

```python
cancel_rate = df.groupby('hotel')['is_canceled'].mean()*100

plt.figure(figsize=(8,5))

sns.barplot(x=cancel_rate.index,
            y=cancel_rate.values)

plt.ylabel("Cancellation Rate (%)")

plt.title("Cancellation Rate by Hotel")

plt.show()
```

```python
df.groupby('hotel').mean(numeric_only=True)

df.groupby('country')['adr'].mean()

df.groupby(['hotel','customer_type']).agg({
'adr':'mean',
'is_canceled':'mean'
})
```
